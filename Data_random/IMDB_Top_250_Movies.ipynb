```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from pylab import rcParams
import pandas as pd
import fastparquet as fp
plt.rcParams["font.family"] = "Arial"
url = '/home/dev/Desktop/Extra coding/Data_random/IMDB_Top_250_Movies.csv'
data = pd.read_csv(url,delimiter = ',')
pd.set_option("display.max_rows", None, "display.max_columns", None)
data.columns = ['rank','name','year','rating','genre','certificate','run_time','tagline','budget','box_office','casts','directors','writers']

# Write the dataframe to a Parquet file
fp.write('example.parquet', data)
# Read the Parquet file into a new dataframe
new_df = fp.ParquetFile('example.parquet').to_pandas()

# Print the new dataframe
# print(new_df)
```

```python
#storing values for each column into a separate list

rank = np.array(new_df.rank)
name = np.array(new_df.name)
year = np.array(new_df.year)
rating = np.array(new_df.rating)
genre = np.array(new_df.genre)
certificate = np.array(new_df.certificate)
run_time = np.array(new_df.run_time)
tagline = np.array(new_df.tagline)
budget = np.array(new_df.budget)
box_office = np.array(new_df.box_office)
casts = np.array(new_df.casts)
directors = np.array(new_df.directors)
writers = np.array(new_df.writers)
```

```python
from aquarel import load_theme

theme = load_theme("boxy_dark")
#https://github.com/lgienapp/aquarel

theme.apply()

# ... plotting code here
fig, ax = plt.subplots(figsize=(20, 20), dpi=80)
# Plot the original scatter plot
plt.scatter(year, genre)
# Filter the data to include only points with 'Western' in the genre
Horror_genre_mask = ['Horror' in g for g in genre]
Horror_year = year[Horror_genre_mask]
Horror_genre = genre[Horror_genre_mask]
# Overplot the points with 'Horror' in the genre
plt.scatter(Horror_year, Horror_genre, color='red', marker='x', s=100, label='Horror Movies')
plt.legend()
plt.show()


theme.apply_transforms()
```
