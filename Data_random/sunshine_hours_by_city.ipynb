```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from pylab import rcParams
import pandas as pd
import fastparquet as fp
import warnings
warnings.filterwarnings("ignore")
from aquarel import load_theme
```

```python
url = '/home/dev/Desktop/Extra coding/Data_random/sunshine_hours_by_city.csv'
data = pd.read_csv(url,delimiter = ',')
pd.set_option("display.max_rows", None, "display.max_columns", None)
data.columns = ['Country','City','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Year']

# Write the dataframe to a Parquet file
fp.write('example.parquet', data)
# Read the Parquet file into a new dataframe
new_df = fp.ParquetFile('example.parquet').to_pandas()

# Print the new dataframe
# print(new_df)
```

```python
#storing values for each column into a separate list
country = np.array(new_df.Country)
city = np.array(new_df.City)
jan = np.array(new_df.Jan)
feb = np.array(new_df.Feb)
mar = np.array(new_df.Mar)
apr = np.array(new_df.City)
may = np.array(new_df.City)
jun = np.array(new_df.Jun)
jul = np.array(new_df.Jul)
aug = np.array(new_df.Aug)
sep = np.array(new_df.Sep)
oct = np.array(new_df.Oct)
nov = np.array(new_df.Nov)
dec = np.array(new_df.Dec)
year = np.array(new_df.Year)
# Country,City,Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov,Dec,Year
```
