```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import pandas as pd
import fastparquet as fp
url = 'https://raw.githubusercontent.com/jovian-explorer/Extra_coding/main/velocity_oscillations.txt'
data = pd.read_csv(url,delimiter = ',')
pd.set_option("display.max_rows", None, "display.max_columns", None)
time= np.linspace(0,10800,240)
# print(len(time))
fig, ax = plt.subplots(figsize=(30,10), dpi=80)
plt.plot(time/60,data)
plt.ylabel("Velocity (in m/sec) $\longrightarrow$")
plt.xlabel("Time (in minutes) $\longrightarrow$")
plt.title("SOLAR PHOTOSPHERIC DOPPLER VELOCITY OSCILLATIONS")
plt.show()
```

```python
frequency = 0.001  # Frequency of the signal
signal = np.sin(2 * np.pi * frequency * time)

# Perform FFT
fft_result = np.fft.fft(signal)

# Calculate the frequencies corresponding to the FFT result
sampling_rate = 1 / (time[1] - time[0])
frequencies = np.fft.fftfreq(len(signal), d=1/sampling_rate)

# Convert time array to minutes
t_minutes = time / 60

# Plot the FFT result
plt.plot(t_minutes, np.abs(fft_result))
plt.xlabel('Time (minutes)')
plt.ylabel('Amplitude')
plt.title('FFT')
plt.show()
```

```python

# Perform FFT
fft_result = np.fft.fft(signal)

# Separate real and imaginary parts
real_values = np.real(fft_result)
imaginary_values = np.imag(fft_result)

# Calculate the frequencies corresponding to the FFT result
sampling_rate = 1 / (time[1] - time[0])
frequencies = np.fft.fftfreq(len(signal), d=1/sampling_rate)

# Plot the real values
plt.plot(frequencies, real_values)
plt.xlabel('Frequency')
plt.ylabel('Amplitude (Real)')
plt.title('FFT (Real Values)')
plt.show()
```
