```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from pylab import rcParams
import pandas as pd

url = '/home/dev/Desktop/hft.txt'
hft = pd.read_csv(url,delimiter = '\t')
pd.set_option("display.max_rows", None, "display.max_columns", None)
hft.columns = ['freq','v_in','v_out']
#storing values for each column into a separate list
X = np.array(hft.freq)
Y = np.array(hft.v_in)
Z = np.array(hft.v_out)

print(hft)
```

```python
plt.plot(X,Y,label = "$V_{in}$")
plt.plot(X,Z,label = "$V_{out}$")
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel("Frequency (in kHz) $\longrightarrow$")
plt.ylabel("Voltage (in V) $\longrightarrow$")
plt.legend()
```

```python
ratio = []
for i in range(len(Y)):
    ratio.append(Y[i]/Z[i])
print(ratio)
```

```python
plt.plot(X,ratio,label = "$V_{in}/V_{out}$")
plt.xlabel("Frequency (in kHz) $\longrightarrow$")
plt.ylabel("Ratio $(V_{in}/V_{out}) \longrightarrow$")
plt.legend()
```
